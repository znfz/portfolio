<center><h3>ArgMax</h3></center>

- Argmax picks the highest value in a classification problem
- Suited for binary classification

```python
import numpy as np

# Initial values for x
x_int = np.array([1,53])
# Labels
class_labels = ["protein", "oil"]

# Use softmax to get the probabilities for each value
x_fin = np.argmax(x_int)

class_labels[x_fin]
```

But what if there are multiple classification solutions and some values are close to each other? 

<center><h3>SoftMax</h3></center>

- Softmax allows to get probabilities for classification problems
- Probabilities add up to 1
- Highest one let's us pick the chances of what output it is

```python
import numpy as np

# Initial values for x
x_int = np.array([1,1.1,1.5,2])
# Labels
class_labels = ["protein", "oil", "rubber","glass"]

# Use softmax to get the probabilities for each value
x_exp = np.exp(x_int)
x_exp_sum = np.sum(x_exp)
x_fin = []

for x in x_exp:
    x_fin.append((x/x_exp_sum)*100)
    
# Step 4: Print probabilities with class labels
probabilities = dict(zip(class_labels, x_fin))

probabilities
```

**Explanation**
- The correct label being a glass has a 40% probability of being correct
- Allows to answer in multiclass classification problem where are possibilities for multiple outputs
